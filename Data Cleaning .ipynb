```python
# Importing registry expression for Data cleaning

import pandas as pd
import re

# Loading CSV
df = pd.read_csv("radiology_data_No_duplicate.csv")

# Removing XXX from the dataset
def clean_text(text):
    text = re.sub(r'\bXXXX\b', '', text) 
    return text.strip()

df['findings'] = df['findings'].apply(clean_text)
df['impression'] = df['impression'].apply(clean_text)

# Saving the cleaned data

df.to_csv("final_cleaned_data.csv", index=False)
```
